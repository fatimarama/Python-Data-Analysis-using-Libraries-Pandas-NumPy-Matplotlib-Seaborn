Q8. Using NumPy, create a function to normalize the 'BloodPressure' values (subtract the mean and divide by the standard deviation).

```python
def normalize_blood_pressure(column_name):
    bp = diabetes[column_name].to_numpy()
    nbp = []
    
    mean_value = np.mean(bp)
    std_value = np.std(bp)
    
    for i in range(len(bp)):
        new_value = (bp[i] - mean_value) / std_value
        nbp.append(new_value)
    return nbp
```

```python
bmi_underweight = 18.5
bmi_normalweight = 24.9
bmi_overweight = 29.9

diabetes['BMI_Class'] = diabetes['BMI'].apply(lambda x: 'Underweight' if x < bmi_underweight
                                             else 'Normal' if x <= bmi_normalweight
                                             else 'Overweight' if x <= bmi_overweight
                                             else 'Obese')
diabetes
```

## Matplotlib Questions

Q1. Plot a histogram of 'Age'.

```python
plt.style.use('default')

age_plot = diabetes["Age"].to_numpy()

plt.figure(figsize=(6, 4))
plt.hist(age_plot, bins=50, color='skyblue', edgecolor="black")
plt.title('Age Histogram')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Q2. Create a scatter plot showing 'Glucose' versus 'Insulin' levels.

```python
glucose = diabetes["Glucose"].to_numpy()
insulin = diabetes["Insulin"].to_numpy()

plt.figure(figsize=(6, 4))
plt.scatter(glucose, insulin, color = "skyblue")
plt.title("Glucose - Insulin Scatter Plot")
plt.xlabel("Glucose")
plt.ylabel("Insulin")
plt.show()
```

Q3. Draw a bar chart comparing the mean 'BMI' across 'Outcome'.

```python
mean_bmi = diabetes.groupby("Outcome")["BMI"].mean()

plt.figure(figsize=(6, 4))
plt.bar(mean_bmi.index, mean_bmi.values, color = "skyblue", edgecolor="black")
plt.title("Bar Chart of mean BMI across Outcome")
plt.xlabel("Outcome")
plt.ylabel("Mean BMI")
plt.show()
```

Q4. Plot a line graph of 'BloodPressure' sorted by 'Age'.

```python
sorted_diabetes_by_age = diabetes.sort_values(by = "Age")

plt.figure(figsize=(10, 4))
plt.plot(sorted_diabetes_by_age["Age"], sorted_diabetes_by_age["BloodPressure"])
plt.title("Line graph of 'BloodPressure' sorted by 'Age'")
plt.xlabel("Sorted Age")
plt.ylabel("Blood Pressure")
plt.show()
```

Q5. Display a box plot of 'BMI' grouped by 'Outcome'.

```python
# if we don't apply .apply(list) here, it will just return an object, so we have to convert it to a list
bmi_outcome = diabetes.groupby("Outcome")["BMI"].apply(list)

# "labels = bmi_outcome.index" we are adding the labels here in the x axis, outcome has 0 and 1 value

plt.figure(figsize=(6, 4))
plt.boxplot(bmi_outcome, labels = bmi_outcome.index)
plt.title("Box plot of 'BMI' grouped by 'Outcome'")
plt.xlabel("Outcome")
plt.ylabel("BMI")
plt.show()

# the plot shows that we have outliers in the dataset
```

Q6. Create a pie chart showing the proportion of diabetes outcomes (0's and 1's).

```python
outcome_counts = diabetes["Outcome"].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(outcome_counts, labels = outcome_counts.index, autopct='%1.2f%%')
plt.title("Proportion of Diabetes Outcomes")
plt.show()
```

Q7. Plot a stacked bar chart of 'Pregnancies' categorized by 'Outcome'.

```python
stacked_diabetes_pregnancies_outcome = diabetes.groupby(['Pregnancies','Outcome']).size().unstack()

stacked_diabetes_pregnancies_outcome.plot(kind='bar', stacked=True, figsize=(6, 4))
plt.title("Stacked bar chart of 'Pregnancies' categorized by 'Outcome'")
plt.xlabel('Pregnancies')
plt.ylabel('Outcome')
plt.show()
```
